```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\Anusha\Documents\Avg growth .csv")
print(data)
data.plot(x = 'Days' ,kind = 'line' )
plt.xticks(rotation=30, horizontalalignment="center")
plt.title("Avg. bean plant growth in moist paper towel")

plt.xlabel('Days of plantation')
plt.ylabel('Growth in cm')
#plt.show()
```

```python

data = pd.read_csv(r"C:\Users\Anusha\Documents\pH new plant.csv")
data.drop(data.filter(regex="Unname"),axis=1, inplace=True)
print(data)
data.plot(x = 'Days' ,kind = 'line' )
plt.xticks(rotation=30, horizontalalignment="center")
plt.title("Avg. bean plant growth watered in pH conc. ")

plt.xlabel('Days of plantation')
plt.ylabel("Growth in cm")
#plt.show()
```

```python
#Rothamsted for the year 1962-1978


crop_yieldr = pd.read_csv(r"C:\Users\Anusha\Documents\YieldCrop(Rothamsted).csv")
print(crop_yieldr)
crop_yieldr.plot(x = 'Plants', kind = 'bar' )

plt.title("Avg. yield of plants in Rothamsted under diff Lime Rates. ")
plt.xticks(rotation=30, horizontalalignment="center")
plt.xlabel("Plants under different Lime rates")
plt.ylabel("Yield in tons per hec. for Rothamsted")
```
